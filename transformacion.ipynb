Ahora que todo los dataframes fueron cargados y creada su respectiva columna de id procedemos a unir las cuatro tabla y asi simplificar la manipulacion de los datasets

```python
stream_platform = pd.concat([amazon, disney, hulu, netflix])
stream_platform.head()
```

```python
# Eliminamos la columna de show_id y ordenamos las demas columnas

stream_platform.drop(columns= 'show_id', inplace=True)
```

```python
# Ordeno las columnas
plat_data = stream_platform.reindex(columns= ['id', 'type', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 'listed_in', 'description', 'platform'])
```

```python
# Reinicio indices
plat_data.reset_index(inplace= True, drop= True)
plat_data.head()
```

```python
# En la columna 'rating' reemplazamos los valores nulos por 'G' el cual corresponde a "general for all audiences"

plat_data['rating'].fillna('G', inplace= True)
plat_data['rating'].unique()
```

```python
# Modifico la columna de fechas para que presente el siguiente formato AAAA-mm-dd

plat_data['date_added'] = pd.to_datetime(plat_data["date_added"])
plat_data['date_added'] = pd.to_datetime(plat_data["date_added"], format= '%Y/%m/%d')
```

```python
plat_data.head()
```

```python
# Cambio los campos de texto a minusculas

plat_data["type"] = plat_data["type"].str.lower()
plat_data["title"] = plat_data["title"].str.lower()
plat_data["director"] = plat_data["director"].str.lower()
plat_data["cast"] = plat_data["cast"].str.lower()
plat_data["country"] = plat_data["country"].str.lower()
plat_data["rating"] = plat_data["rating"].str.lower()
plat_data["listed_in"] = plat_data["listed_in"].str.lower()
plat_data["description"] = plat_data["description"].str.lower()
plat_data["duration"] = plat_data["duration"].str.lower()
```

6 - El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
# Convirtiendo la columna "duration" en "duration_int y duration_type, conteniendo valores integer y string respectivamente
# este ultimo indica el tipo de duracion "min" y "season"

plat_data[['duration_int', 'duration_type']] = plat_data['duration'].str.extract('(\d+)\s*(\w+)')
plat_data['duration_type'].replace('seasons', 'season', inplace= True)
```

```python
plat_data['duration_type'].unique()
```

Ahora que las tablas de las plataformas estan transformadas y unidas, procederemos a trabajar con las tablas de rating.

 - El trabajo consistira en concatenar cada una de las tablas, transformar el formato de fecha y del resultado final creare un dataframe con el promedio de score por pelicula que nos servira para fusionarla con la tabla de plataformas.

```python
# Cargamos los csv a pandas
rating1 = pd.read_csv('ratings/1.csv')
rating2 = pd.read_csv('ratings/2.csv')
rating3 = pd.read_csv('ratings/3.csv')
rating4 = pd.read_csv('ratings/4.csv')
rating5 = pd.read_csv('ratings/5.csv')
rating6 = pd.read_csv('ratings/6.csv')
rating7 = pd.read_csv('ratings/7.csv')
rating8 = pd.read_csv('ratings/8.csv')

# Concateno los csv y reinicio los indices
rating_data = pd.concat((rating1, rating2, rating3, rating4, rating5, rating6, rating7, rating8))
rating_data.reset_index(inplace= True, drop= True)
rating_data.head()


# Cambio el nombre de la columna rating por score y movieId por id para mantener uniformidad

rating_data.rename(columns= {'rating' : 'score'}, inplace= True)
rating_data.rename(columns= {'movieId' : 'id'}, inplace= True)
rating_data['userid'] = rating_data['userId']
rating_data.drop(['userId'], axis= 1, inplace=True)
```

```python
rating_data.head()
```

```python
# Crear columna date el cual correspondera a la fecha 
rating_data['date'] =  pd.to_datetime(rating_data['timestamp'], unit='s').dt.strftime('%Y-%m-%d')
```

```python
#Elimino la columna timestamp

rating_data.drop(columns= 'timestamp', inplace= True)
```

```python
# Creo una tabla con el promedio de score agrupado por peliculas

score_movie = rating_data[['id', 'score']].groupby('id').mean()

score_movie
```

```python
# Ahora procedo a combinar los dataframes "plat_data" y "score_movie" utilizando merge por medio de "id"

movie_data = pd.merge(plat_data, score_movie, how='left', on='id')
```

```python
# Creo un dataframe con el nombre de las peliculas y sus respectivos id
movie_names = movie_data[['id', 'title']]
movie_names.head()
```

```python
# Combinamos el dataset de rating con el dataset de plataformas para tener el nombre de las peliculas

rating_data = pd.merge(rating_data, movie_names , how= 'left', on='id')
rating_data.shape
```
